```python
X_mean=X.iloc[:,0:25].mean(axis=0)
X_std=X.iloc[:,0:25].std(axis=0)
X_normalized=(X.iloc[:,0:25]-X_mean)/X_std
```

```python
X_normalized
```

```python
split_index=int(0.8*len(X))
X_train,X_test=X_normalized.iloc[0:split_index,:],X_normalized.iloc[split_index:len(X),:]
Y_train,Y_test=X.iloc[0:split_index,25],X.iloc[split_index:len(X),25]
```

```python
def initialize_weights(n_features):
    w=np.zeros(n_features)
    b=0
    return w,b
```

```python
def predict(X,w,b):
    return X.dot(w)+b
```

```python
def compute_cost(X,Y,w,b):
    m=len(Y)
    predictions=predict(X,w,b)
    cost=(1/(2*m))*np.sum((predictions-Y)**2)
    return cost
```

```python
def gradient_descent(X,Y,w,b,alpha,iterations):
    m=len(Y)
    cost_history=[]

    for i in range(iterations):
        predictions=predict(X,w,b)
        dw=(1/m)*X.T.dot(predictions-Y)
        db=(1/m)*np.sum(predictions-Y)
        w-=alpha*dw
        b-=alpha*db
        cost=compute_cost(X,Y,w,b)
        cost_history.append(cost)
        if i%100==0:
            print(f"Iteration {i}/{iterations},Cost: {cost}")

    return w,b,cost_history
```

```python
alpha=0.01
iterations=1000
w,b=initialize_weights(X_train.shape[1])
w,b,cost_history=gradient_descent(X_train,Y_train,w,b,alpha,iterations)
print("Trained weights:",w)
print("Trained bias:",b)
```

```python
plt.plot(range(iterations), cost_history)
plt.xlabel("Iterations")
plt.ylabel("Cost (MSE)")
plt.title("Cost Function Convergence")
plt.show()
```
