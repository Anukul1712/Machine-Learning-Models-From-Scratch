```python
Y_binary=data.iloc[:,-2].values
Y_multi=data.iloc[:,-1].values
```

```python
def one_hot_encode(Y,num_classes):
    one_hot=np.zeros((Y.size,num_classes))
    one_hot[np.arange(Y.size),Y-1]=1  # Assuming classes are 1-indexed
    return one_hot

Y_multi_one_hot=one_hot_encode(Y_multi,10)
```

```python
split_index=int(0.8*len(X))
X_train,X_test=X[0:split_index,:],X[split_index:len(X),:]
Y_binary_train,Y_binary_test=Y_binary[0:split_index,],Y_binary[split_index:len(X),]
Y_multi_train,Y_multi_test=Y_multi_one_hot[0:split_index,],Y_multi_one_hot[split_index:len(X),]
```

```python
input_size=1024
hidden_size1=512
output_size_binary=1
output_size_multi=10
```

```python
np.random.seed(42)
w={
    "W1":np.random.randn(input_size,hidden_size1).astype(np.float32)*np.sqrt(2.0/input_size),
    "b1":np.zeros((1,hidden_size1),dtype=np.float32),
    "W_binary":np.random.randn(hidden_size1,output_size_binary).astype(np.float32)*np.sqrt(2.0/hidden_size1),
    "b_binary":np.zeros((1,output_size_binary),dtype=np.float32),
    "W_multi":np.random.randn(hidden_size1, output_size_multi).astype(np.float32)*np.sqrt(2.0/hidden_size1),
    "b_multi":np.zeros((1,output_size_multi),dtype=np.float32)
}
```

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

```python
def softmax(z):
    exp_z=np.exp(z-np.max(z,axis=1,keepdims=True))
    return exp_z/np.sum(exp_z,axis=1,keepdims=True)
```

```python
def relu(z):
    return np.maximum(0,z)
```

```python
def forward_propagation(X,w):
    Z1=np.dot(X,w["W1"])+w["b1"]
    A1=relu(Z1)
    
    Z_binary=np.dot(A1,w["W_binary"])+w["b_binary"]
    A_binary=sigmoid(Z_binary)
    
    Z_multi=np.dot(A1,w["W_multi"])+w["b_multi"]
    A_multi=softmax(Z_multi)
    
    cache={"Z1":Z1,"A1":A1,"Z_binary":Z_binary,"A_binary":A_binary,"Z_multi":Z_multi,"A_multi":A_multi}
    return A_binary,A_multi,cache
```

```python
def compute_cost(A_binary,Y_binary,A_multi,Y_multi_one_hot):
    m=Y_binary.shape[0]
    binary_cost=-np.mean(Y_binary*np.log(A_binary+1e-8)+(1-Y_binary)*np.log(1-A_binary+1e-8))
    multi_cost=-np.mean(np.sum(Y_multi_one_hot*np.log(A_multi+1e-8),axis=1))
    return binary_cost,multi_cost
```

```python
def backward_propagation(X,Y_binary,Y_multi_one_hot,cache,w):
    m=X.shape[0]
    dZ_binary=cache["A_binary"]-Y_binary.reshape(-1,1)
    dW_binary=np.dot(cache["A1"].T,dZ_binary)/m
    db_binary=np.sum(dZ_binary,axis=0,keepdims=True)/m
    
    dZ_multi=cache["A_multi"]-Y_multi_one_hot
    dW_multi=np.dot(cache["A1"].T,dZ_multi)/m
    db_multi=np.sum(dZ_multi,axis=0,keepdims=True)/m
    
    dA1=np.dot(dZ_binary,w["W_binary"].T)+np.dot(dZ_multi,w["W_multi"].T)
    dZ1=dA1*(cache["Z1"]>0)
    dW1=np.dot(X.T,dZ1)/m
    db1=np.sum(dZ1,axis=0,keepdims=True)/m
    
    gradients={"dW1":dW1,"db1":db1,"dW_binary":dW_binary,"db_binary":db_binary,"dW_multi":dW_multi,"db_multi":db_multi}
    return gradients
```

```python
def update_weights(w,gradients,learning_rate,beta1,beta2,epsilon,t,v,s):
    for key in w.keys():
        if key.startswith("W") or key.startswith("b"):
            v[key]=beta1*v[key]+(1-beta1)*gradients["d"+key]
            s[key]=beta2*s[key]+(1-beta2)*(gradients["d"+key]**2)
            v_corrected=v[key]/(1-beta1**t)
            s_corrected=s[key]/(1-beta2**t)
            w[key]-=learning_rate*v_corrected/(np.sqrt(s_corrected)+epsilon)
    return w,v,s
```

```python
def mini_batch_generator(X,Y_binary,Y_multi_one_hot,batch_size):
    m=X.shape[0]
    indices=np.random.permutation(m)
    X_shuffled=X[indices]
    y_binary_shuffled=Y_binary[indices]
    y_multi_shuffled=Y_multi_one_hot[indices]
    for i in range(0,m,batch_size):
        end=i+batch_size
        yield X_shuffled[i:end],y_binary_shuffled[i:end],y_multi_shuffled[i:end]
```

```python
def calculate_accuracy(Y_true,Y_pred):
    if Y_pred.shape[1]==1:
        Y_pred_class=(Y_pred>0.5).astype(int)
        return np.mean(Y_pred_class==Y_true.reshape(-1,1))
    else:
        Y_pred_class=np.argmax(Y_pred,axis=1)+1
        return np.mean(Y_pred_class==Y_true)
```

```python
def train_model(X, Y_binary, Y_multi, w, epochs=30, batch_size=64, initial_learning_rate=0.001, beta1=0.9, beta2=0.999, epsilon=1e-8):
    v = {key: np.zeros_like(val) for key, val in w.items()}
    s = {key: np.zeros_like(val) for key, val in w.items()}

    learning_rate = initial_learning_rate

    for epoch in range(1, epochs + 1):
        epoch_binary_cost = 0
        epoch_multi_cost = 0
        num_batches = 0

        for X_batch, y_binary_batch, y_multi_batch_one_hot in mini_batch_generator(X, Y_binary, Y_multi, batch_size):
            A_binary, A_multi, cache = forward_propagation(X_batch, w)
            binary_cost, multi_cost = compute_cost(A_binary, y_binary_batch, A_multi, y_multi_batch_one_hot)
            gradients = backward_propagation(X_batch, y_binary_batch, y_multi_batch_one_hot, cache, w)
            w, v, s = update_weights(w, gradients, learning_rate, beta1, beta2, epsilon, epoch, v, s)

            epoch_binary_cost += binary_cost
            epoch_multi_cost += multi_cost
            num_batches += 1

        epoch_binary_cost /= num_batches
        epoch_multi_cost /= num_batches

        # Compute accuracy on the training set
        train_A_binary, train_A_multi, _ = forward_propagation(X_train, w)
        train_binary_accuracy = calculate_accuracy(Y_binary_train, train_A_binary)
        train_multi_accuracy = calculate_accuracy(np.argmax(Y_multi_train, axis=1) + 1, train_A_multi)

        # Compute accuracy on the test set
        test_A_binary, test_A_multi, _ = forward_propagation(X_test, w)
        test_binary_accuracy = calculate_accuracy(Y_binary_test, test_A_binary)
        test_multi_accuracy = calculate_accuracy(np.argmax(Y_multi_test, axis=1) + 1, test_A_multi)

        
        print(f"Epoch {epoch}: Binary Cost = {epoch_binary_cost:.4f}, Multi-class Cost = {epoch_multi_cost:.4f}, \
        Train Binary Accuracy = {train_binary_accuracy:.4f}, Train Multi-class Accuracy = {train_multi_accuracy:.4f}, \
        Test Binary Accuracy = {test_binary_accuracy:.4f}, Test Multi-class Accuracy = {test_multi_accuracy:.4f}")

    return w
```

```python
trained_weights = train_model(X_train, Y_binary_train, Y_multi_train, w, epochs=30, batch_size=64, initial_learning_rate=0.001, beta1=0.9, beta2=0.999, epsilon=1e-8)
```

```python
test_A_binary, test_A_multi, _ = forward_propagation(X_test, trained_weights)
test_binary_accuracy = calculate_accuracy(Y_binary_test, test_A_binary)
test_multi_accuracy = calculate_accuracy(np.argmax(Y_multi_test, axis=1) + 1, test_A_multi)

print(f"Final Test Binary Accuracy: {test_binary_accuracy:.4f}")
print(f"Final Test Multi-class Accuracy: {test_multi_accuracy:.4f}")
```
