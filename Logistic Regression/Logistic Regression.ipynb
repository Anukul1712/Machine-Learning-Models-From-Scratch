```python
split_index=int(0.8 * len(data))
end=len(data)
X_train,X_test=data.iloc[0:split_index,0:20],data.iloc[split_index:end,0:20]
Y_train,Y_test=data.iloc[0:split_index,20],data.iloc[split_index:end,20]
X_train.shape
```

```python
def predict(x,w,b):
    z=np.dot(x,w)+b
    return z
```

```python
def sigmoid(z):
    g=1/(1+np.exp(-z))
    return g
```

```python
def initialize_weights(n_features):
    w=np.zeros(n_features)
    b=0
    return w,b
```

```python
def compute_cost(x,y,w,b):
    m,n=x.shape
    z=predict(x,w,b)
    g=sigmoid(z)
    total_cost=-(1/m)*np.sum(y*np.log(g+1e-10)+(1-y)*np.log(1-(g+1e-10)))
    return total_cost
```

```python
def gradient_descent(x,y,w,b,alpha,iterations):
    m=len(y)
    cost_history=[]
    for i in range(iterations):
        z=predict(x,w,b)
        g=sigmoid(z)
        dw=(1/m)*np.dot(x.T,g-y)
        db=(1/m)*np.sum(g-y)
        w-=alpha*dw
        b-=alpha*db
        cost=compute_cost(x,y,w,b)
        cost_history.append(cost)
        if i%100==0:
            print(f"Iteration {i}/{iterations},Cost: {cost}")
            
    return w,b,cost_history
```

```python
alpha=0.1
iterations=1000

w,b=initialize_weights(X_train.shape[1])

w,b,cost_history=gradient_descent(X_train,Y_train,w,b,alpha,iterations)

print("Trained weights:",w)
print("Trained bias:",b)
```

```python
plt.plot(range(len(cost_history)), cost_history)
plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.title('Cost Function Convergence')
plt.show()
```

```python
def predict_classes(X, W, b, threshold=0.5):
    Z = np.dot(X, W) + b
    A = sigmoid(Z)
    return (A >= threshold).astype(int)
```

```python
def accuracy(y_true, y_pred):
    return np.mean(y_pred == y_true)
```

```python
y_pred_train = predict_classes(X_train, w, b)
y_pred_test = predict_classes(X_test, w, b)
train_accuracy = accuracy(Y_train, y_pred_train)*100
test_accuracy = accuracy(Y_test, y_pred_test)*100

print(f"Train Accuracy: {train_accuracy}")
print(f"Test Accuracy: {test_accuracy}")
```

```python
def calculate_precision(y_true, y_pred):
    tp = np.sum((y_true==1) & (y_pred==1))
    fp = np.sum((y_true==0) & (y_pred==1))
    return (tp / (tp+fp)) if (tp+fp) != 0 else 0
```

```python
precision_test = calculate_precision(Y_test, y_pred_test)
print(f"Precision for Test Set: {precision_test}")
precision_train = calculate_precision(Y_train, y_pred_train)
print(f"Precision for Training Set: {precision_train}")
```

```python
def calculate_recall(y_true,y_pred):
    tp = np.sum((y_true==1)&(y_pred==1))
    fn = np.sum((y_true==1)&(y_pred==0))
    return (tp/(tp+fn)) if (tp+fn)!=0 else 0
```

```python
recall_test=calculate_recall(Y_test,y_pred_test)
print(f"Recall for Test Set: {recall_test}")
recall_train=calculate_recall(Y_train,y_pred_train)
print(f"Recall for Training Set: {recall_train}")
```

```python
def calculate_f1score(precision,recall):
    return (2*precision*recall)/(precision+recall)
```

```python
f1score_train=calculate_f1score(precision_train,recall_train)
print(f"f1 for Training Set: {f1score_train}")
```

```python
f1score_test=calculate_f1score(precision_test,recall_test)
print(f"f1 for Test Set: {f1score_test}")
```
