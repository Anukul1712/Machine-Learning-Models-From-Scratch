```python
X_mean=data.iloc[:,0:20].mean(axis=0)
X_std=data.iloc[:,0:20].std(axis=0)
X_normalized=(data.iloc[:,0:20]-X_mean)/X_std
```

```python
X_normalized
```

```python
split_index=int(0.8*len(data))
X_train,X_test=X_normalized.iloc[0:split_index,:].to_numpy(),X_normalized.iloc[split_index:len(data),:].to_numpy()
Y_train,Y_test=data.iloc[0:split_index,20].to_numpy(),data.iloc[split_index:len(data),20].to_numpy()
```

```python
Y_train
```

```python
def frequent_labels(labels):
    unique_labels, counts = np.unique(labels, return_counts=True)
    return unique_labels[np.argmax(counts)]
```

```python
def batch_knn(X_train, Y_train, X_test, k, batch_size=1000):
    predictions=[]
    for start in range(0, len(X_test), batch_size):
        end = min(start + batch_size, len(X_test))
        X_test_batch = X_test[start:end]
        
        distances = np.linalg.norm(X_test_batch[:, None] - X_train, axis=2)
        
        nearest_indices = np.argpartition(distances, k, axis=1)[:, :k]
        nearest_labels = Y_train[nearest_indices]
        
        batch_predictions = np.apply_along_axis(frequent_labels, 1, nearest_labels)
        predictions.extend(batch_predictions)
    
    return np.array(predictions)
```

```python
k = 5
batch_size = 500

Y_pred = batch_knn(X_train, Y_train, X_test, k, batch_size=batch_size)

accuracy = np.mean(Y_pred == Y_test) * 100
print(f"Test Accuracy: {accuracy:.2f}%")
```
